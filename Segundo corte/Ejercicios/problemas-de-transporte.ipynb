## Ejemplo 1: Formulación de Powerco

Powerco tiene tres plantas de energía eléctrica que abastecen las necesidades de cuatro ciudades. Cada planta puede suministrar las siguientes cantidades de kilovatios-hora (kwh) de electricidad: planta 1—35 millones; planta 2—50 millones; planta 3—40 millones (ver Tabla 1). Las demandas máximas de energía en estas ciudades, que ocurren al mismo tiempo (2 p.m.), son las siguientes (en kwh): ciudad 1—45 millones; ciudad 2—20 millones; ciudad 3—30 millones; ciudad 4—30 millones. Los costos de enviar 1 millón de kwh de electricidad desde una planta a una ciudad dependen de la distancia que la electricidad debe recorrer. Formula un problema de programación lineal (PL) para minimizar el costo de satisfacer la demanda máxima de energía de cada ciudad.

### Tabla 1

|----------|------------|------------|------------|------------|-------------------------------|
| Planta 1 | $8         | $6         | $10        | $9         | 35                            |
| Planta 2 | $9         | $12        | $13        | $7         | 50                            |
| Planta 3 | $14        | $9         | $16        | $5         | 40                            |
| **Demanda (millones de kwh)** | **45** | **20**    | **30**    | **30**   |                               |


### Paso a Paso para Modelar el Problema de Powerco

#### 1. Definir las Variables de Decisión

Definimos las variables de decisión como la cantidad de kilovatios-hora (kwh) de electricidad que se enviará desde cada planta a cada ciudad:

* $x_{ij}$: Cantidad de electricidad (en millones de kwh) enviada desde la planta $i$ a la ciudad $j$.

Donde:

* $i = 1, 2, 3$ representa las plantas (Planta 1, Planta 2, Planta 3).
* $j = 1, 2, 3, 4$ representa las ciudades (Ciudad 1, Ciudad 2, Ciudad 3, Ciudad 4).

#### 2. Definir la Función Objetivo

La función objetivo es **minimizar el costo total de envío de electricidad** desde las plantas a las ciudades:

$$\text{Minimizar } Z = 8x_{11} + 6x_{12} + 10x_{13} + 9x_{14} + 9x_{21} + 12x_{22} + 13x_{23} + 7x_{24} + 14x_{31} + 9x_{32} + 16x_{33} + 5x_{34}$$

donde:

* $c_{ij}$ es el costo de enviar 1 millón de kwh de electricidad desde la planta $i$ a la ciudad $j$, de acuerdo con la **Tabla 1**:
    * Planta 1 a Ciudad 1: $8
    * Planta 1 a Ciudad 2: $6
    * Planta 1 a Ciudad 3: $10
    * Planta 1 a Ciudad 4: $9
    * Planta 2 a Ciudad 1: $9
    * Planta 2 a Ciudad 2: $12
    * Planta 2 a Ciudad 3: $13
    * Planta 2 a Ciudad 4: $7
    * Planta 3 a Ciudad 1: $14
    * Planta 3 a Ciudad 2: $9
    * Planta 3 a Ciudad 3: $16
    * Planta 3 a Ciudad 4: $5

#### 3. Establecer las Restricciones

##### a) Restricciones de Suministro (Capacidad de las Plantas)

Cada planta tiene una capacidad máxima de suministro:

$$
x_{11} + x_{12} + x_{13} + x_{14} \leq 35 \quad (\text{Capacidad de Planta 1})\\
x_{21} + x_{22} + x_{23} + x_{24} \leq 50 \quad (\text{Capacidad de Planta 2})\\
x_{31} + x_{32} + x_{33} + x_{34} \leq 40 \quad (\text{Capacidad de Planta 3})\\
$$

##### b) Restricciones de Demanda (Requerimientos de las Ciudades)

Cada ciudad tiene una demanda fija que debe ser satisfecha:

$$
x_{11} + x_{21} + x_{31} = 45 \quad (\text{Demanda de Ciudad 1})\\
x_{12} + x_{22} + x_{32} = 20 \quad (\text{Demanda de Ciudad 2})\\
x_{13} + x_{23} + x_{33} = 30 \quad (\text{Demanda de Ciudad 3})\\
x_{14} + x_{24} + x_{34} = 30 \quad (\text{Demanda de Ciudad 4})
$$

##### c) Restricciones de No Negatividad

No se puede enviar una cantidad negativa de electricidad:

$$x_{ij} \geq 0 \quad \forall i, j$$

#### 4. Formulación del Modelo Completo

El modelo de programación lineal (PL) para este problema de transporte se formula de la siguiente manera:

$$
\text{Minimizar } Z = 8x_{11} + 6x_{12} + 10x_{13} + 9x_{14} + 9x_{21} + 12x_{22} + 13x_{23} + 7x_{24} + 14x_{31} + 9x_{32} + 16x_{33} + 5x_{34}
$$

Sujeto a:

1. Restricciones de Suministro:

$$
x_{11} + x_{12} + x_{13} + x_{14} \leq 35\\
x_{21} + x_{22} + x_{23} + x_{24} \leq 50\\
x_{31} + x_{32} + x_{33} + x_{34} \leq 40
$$

2. Restricciones de Demanda:

$$
x_{11} + x_{21} + x_{31} = 45\\
x_{12} + x_{22} + x_{32} = 20\\
x_{13} + x_{23} + x_{33} = 30\\
x_{14} + x_{24} + x_{34} = 30
$$

3. Restricciones de No Negatividad:

$$
x_{ij} \geq 0 \quad \forall i, j
$$

````python
#@title Grafo
import matplotlib.pyplot as plt

# Definir las posiciones de las plantas (suministro) y ciudades (demanda)
plantas = {'Planta 1': (-1, 2), 'Planta 2': (-1, 1), 'Planta 3': (-1, 0)}
capacidades = {'Planta 1': 35, 'Planta 2': 50, 'Planta 3': 40}  # Oferta de cada planta

ciudades = {'Ciudad 1': (1, 2.5), 'Ciudad 2': (1, 1.5), 'Ciudad 3': (1, 0.5), 'Ciudad 4': (1, -0.5)}
demandas = {'Ciudad 1': 45, 'Ciudad 2': 20, 'Ciudad 3': 30, 'Ciudad 4': 30}  # Demanda de cada ciudad

# Crear el gráfico
figex1, axex1 = plt.subplots()

# Dibujar las plantas (nodos de suministro)
for planta, (x, y) in plantas.items():
    axex1.plot(x, y, 'bo', markersize=10)  # Dibujar nodos en azul
    axex1.text(x - 0.2, y, f"{planta}\n(Oferta: {capacidades[planta]})", fontsize=10, ha='right')  # Etiquetas de los nodos con oferta

# Dibujar las ciudades (nodos de demanda)
for ciudad, (x, y) in ciudades.items():
    axex1.plot(x, y, 'ro', markersize=10)  # Dibujar nodos en rojo
    axex1.text(x + 0.2, y, f"{ciudad}\n(Demanda: {demandas[ciudad]})", fontsize=10, ha='left')  # Etiquetas de los nodos con demanda

# Dibujar las flechas (arcos) entre todas las plantas y ciudades (posibles envíos)
for px, py in plantas.values():
    for cx, cy in ciudades.values():
        axex1.arrow(px, py, cx - px - 0.2, cy - py, head_width=0.1, head_length=0.2, fc='gray', ec='gray', linestyle='--')

# Configurar el gráfico
axex1.set_title("Problema de Transporte de Powerco (Sin Solución)", fontsize=14)
axex1.set_xlim(-2, 2)
axex1.set_ylim(-1, 3)
axex1.axis('off')  # Ocultar los ejes

plt.show()
````

````python
from ortools.linear_solver import pywraplp

# Crear el solver
solver = pywraplp.Solver.CreateSolver('GLOP')

# Definir parámetros del problema
# Costos de envío desde cada planta a cada ciudad
cost = [
    [8, 6, 10, 9],   # Costos desde Planta 1 a Ciudad 1, 2, 3, 4
    [9, 12, 13, 7],  # Costos desde Planta 2 a Ciudad 1, 2, 3, 4
    [14, 9, 16, 5]   # Costos desde Planta 3 a Ciudad 1, 2, 3, 4
]

# Suministro de cada planta
supply = [35, 50, 40]  # Capacidad de Planta 1, 2, 3

# Demanda de cada ciudad
demand = [45, 20, 30, 30]  # Demanda de Ciudad 1, 2, 3, 4

# Definir variables de decisión
x = {}  # Diccionario para almacenar variables de decisión

for i in range(len(supply)):
    for j in range(len(demand)):
        # Crear variable x[i,j] que representa el envío de electricidad desde planta i a ciudad j
        x[(i, j)] = solver.NumVar(0, solver.infinity(), f'x[{i},{j}]')

# Definir la función objetivo (minimizar costo total)
objective = solver.Objective()
for i in range(len(supply)):
    for j in range(len(demand)):
        objective.SetCoefficient(x[(i, j)], cost[i][j])
objective.SetMinimization()

# Agregar restricciones de suministro (capacidad de plantas)
for i in range(len(supply)):
    constraint = solver.Constraint(0, supply[i])  # Cada planta no puede suministrar más de su capacidad
    for j in range(len(demand)):
        constraint.SetCoefficient(x[(i, j)], 1)

# Agregar restricciones de demanda (requerimientos de ciudades)
for j in range(len(demand)):
    constraint = solver.Constraint(demand[j], demand[j])  # Cada ciudad debe recibir su demanda exacta
    for i in range(len(supply)):
        constraint.SetCoefficient(x[(i, j)], 1)

# Resolver el problema
status = solver.Solve()

# Verificar el estado de la solución
if status == pywraplp.Solver.OPTIMAL:
    print('Solución óptima encontrada:')
    print('Costo total =', solver.Objective().Value())
    for i in range(len(supply)):
        for j in range(len(demand)):
            print(f'Cantidad enviada desde Planta {i+1} a Ciudad {j+1}: {x[(i, j)].solution_value()} millones de kwh')
else:
    print('No se encontró una solución óptima.')
````

````python
from gurobipy import Model, GRB

# Definir parámetros del problema
# Costos de envío desde cada planta a cada ciudad
cost = [
    [8, 6, 10, 9],   # Costos desde Planta 1 a Ciudad 1, 2, 3, 4
    [9, 12, 13, 7],  # Costos desde Planta 2 a Ciudad 1, 2, 3, 4
    [14, 9, 16, 5]   # Costos desde Planta 3 a Ciudad 1, 2, 3, 4
]

# Suministro de cada planta
supply = [35, 50, 40]  # Capacidad de Planta 1, 2, 3

# Demanda de cada ciudad
demand = [45, 20, 30, 30]  # Demanda de Ciudad 1, 2, 3, 4

# Crear el modelo de optimización
model = Model('Problema de Transporte Powerco')

# Crear variables de decisión
x = {}  # Diccionario para almacenar las variables

for i in range(len(supply)):
    for j in range(len(demand)):
        # Crear variable x[i,j] que representa la cantidad enviada desde planta i a ciudad j
        x[(i, j)] = model.addVar(vtype=GRB.CONTINUOUS, name=f'x_{i}_{j}')

# Definir la función objetivo (minimizar costo total)
model.setObjective(
    sum(cost[i][j] * x[(i, j)] for i in range(len(supply)) for j in range(len(demand))),
    GRB.MINIMIZE
)

# Agregar restricciones de suministro (capacidad de las plantas)
for i in range(len(supply)):
    model.addConstr(
        sum(x[(i, j)] for j in range(len(demand))) <= supply[i],
        name=f'Suministro_Planta_{i}'
    )

# Agregar restricciones de demanda (requerimientos de las ciudades)
for j in range(len(demand)):
    model.addConstr(
        sum(x[(i, j)] for i in range(len(supply))) == demand[j],
        name=f'Demanda_Ciudad_{j}'
    )

# Optimizar el modelo
model.optimize()

# Verificar el estado de la solución
if model.status == GRB.OPTIMAL:
    print('Solución óptima encontrada:')
    print('Costo total =', model.objVal)
    for i in range(len(supply)):
        for j in range(len(demand)):
            print(f'Cantidad enviada desde Planta {i+1} a Ciudad {j+1}: {x[(i, j)].x} millones de kwh')
else:
    print('No se encontró una solución óptima.')
````
